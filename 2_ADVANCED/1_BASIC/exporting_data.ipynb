```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
datajobs = dataset['train'].to_pandas()

# Data Cleanup
datajobs['job_posted_date'] = pd.to_datetime(datajobs['job_posted_date'])
```

```python
datajobs[pd.notna(datajobs["salary_year_avg"])] ["salary_year_avg"]
```

```python
datajobs.apply(datajobs["salary_year_avg"])
```

```python
new_data = datajobs [pd.notna(datajobs["salary_year_avg"])].copy()
```

```python
def projected_salary(salary):
    return salary * 1.03

new_data["salary_year_inflated"] = new_data["salary_year_avg"].apply(func = projected_salary)

new_data [["salary_year_avg", "salary_year_inflated"]].head(5)
```

```python
new_data ['salary_year_inflated'] = new_data ["salary_year_avg"].apply(lambda salary: salary * 1.03)

new_data [["salary_year_avg", "salary_year_inflated"]].head(5)
```

```python
import ast

ast.literal_eval(datajobs ["job_skills"][1])

def conversion_skills(skills):
    if pd.notna(skills):
        return ast.literal_eval(skills)

datajobs ["job_skills"] = datajobs['job_skills'].apply(conversion_skills)
```

```python
def projected_salary(row):
    if "Senior" in row["job_title_short"]:
        return row["salary_year_avg"] * 1.05
    else: 
        return row["salary_year_avg"] * 1.03

new_data ["salary_year_inflated"] = new_data.apply(projected_salary, axis = 1)

new_data [["job_title_short", "salary_year_avg", "salary_year_inflated"]].head(20)
```
