```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python

#filtering for data analysts in the UK
datajobsUS= df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

datajobsUS= datajobsUS.dropna(subset=["salary_year_avg"])
```

```python
datajobsUS = datajobsUS.explode("job_skills")

datajobsUS[["salary_year_avg", "job_skills"]]
```

```python
datajobsUS_group = datajobsUS.groupby("job_skills") ["salary_year_avg"].agg(["count", "median"])

datajobsUS_top_pay = datajobsUS_group.sort_values(by = "median", ascending = False).head(10)

```
