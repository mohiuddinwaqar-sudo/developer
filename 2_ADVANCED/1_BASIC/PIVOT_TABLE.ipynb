```python
import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#loading dataset 

dataset = load_dataset("lukebarousse/data_jobs")

data_jobs = dataset['train'].to_pandas()

data_jobs ["job_posted_date"] = pd.to_datetime(data_jobs.job_posted_date)
```

```python
data_jobs.pivot_table(values = "salary_year_avg",index = "job_title_short", columns = "job_country", aggfunc = 'median')
```

```python
data_jobs.groupby("job_title_short") ['salary_year_avg'].agg('median')


```

```python
top_countries = data_jobs["job_country"].value_counts().head(6).index
top_countries
```

```python
datajobs_country_salary = data_jobs.pivot_table(values = "salary_year_avg",
                                                index = "job_country", 
                                                columns = "job_title_short",
                                                aggfunc = "median"
)

datajobs_country_salary = datajobs_country_salary.loc[top_countries]

job_titles = ["Data Analyst", "Data Engineer", "Data Scientist"]

datajobs_country_salary = datajobs_country_salary[job_titles]

datajobs_country_salary
```
