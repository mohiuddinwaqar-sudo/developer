```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python

import seaborn as sns
```

```python
datajobsaus = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

datajobsaus = datajobsaus.dropna(subset=["salary_year_avg"])

sns.displot(
    datajobsaus["salary_year_avg"],
    bins=30,
    kde=True,                                # ðŸ‘ˆ adds the smooth curve
    fill = True,
    color="green"
)

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f"${int(x/1000)}K"))

plt.tight_layout()
```

```python
sns.displot(
    datajobsaus["salary_year_avg"],
    kind = 'kde',                                # ðŸ‘ˆ adds the smooth curve
    fill = True,
    color="green"
)

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f"${int(x/1000)}K"))

plt.tight_layout()
```
