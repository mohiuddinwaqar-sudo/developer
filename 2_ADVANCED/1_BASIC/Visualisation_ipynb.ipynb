```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
job_salary = data_jobs.groupby('job_title_short')["salary_year_avg"].agg("median").sort_values()
```

```python
job_salary.plot(kind="barh", color = "red")
plt.xticks(rotation = 60, ha = 'right')
plt.title('Job Salary By Position', weight = 'bold')
plt.xlabel('Job Title', weight = 'bold')
plt.ylabel('')
plt.show()
```
